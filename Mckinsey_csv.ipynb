```python
# from df, find the average TOTAL gdp of iceland the year 2007

df3 = df.query('country == "Iceland" and year < 2000')
df3["total_gdp"] = df3['population'] * df3['gdp_cap']
```

```python
df3 = df.query('country == "Iceland" and year < 2000') .copy() #here we make a copy to avoid above warning error
df3["total_gdp"] = df3['population'] * df3['gdp_cap']
```
